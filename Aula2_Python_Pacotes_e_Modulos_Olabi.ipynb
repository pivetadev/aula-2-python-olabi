**O que seria esse tal de Def?**
Vamos fazer um bolo. Em vez de misturar todos os ingredientes manualmente toda vez que quiser fazer um bolo, voc pode seguir uma receita. A receita  como uma funo em Python: um conjunto de instrues que voc pode seguir sempre que precisar.

**Como Criar uma Funo**
Para criar uma funo em Python, usamos a palavra-chave def, que significa "definir". Vamos seguir um exemplo bsico:

````python
# Criar o mdulo chamado matematica.py
with open('matematica.py', 'w') as f:
    f.write("""
def soma(a, b):
    return a + b

def subtracao(a, b):
    return a - b

def multiplicacao(a, b):
    return a * b

def divisao(a, b):
    if b == 0:
        return "No  possvel dividir por zero!"
    return a / b
    """)
````

**Como Funciona o with**

Quando voc usa with, voc est dizendo ao Python para garantir que um recurso seja corretamente adquirido e, em seguida, liberado, mesmo que ocorra um erro durante o uso desse recurso. O with simplifica o cdigo e torna a liberao de recursos mais segura.

`with:` Inicia um gerenciador de contexto para o arquivo. Isso garante que o arquivo ser fechado corretamente aps a operao, mesmo que ocorra um erro.
open
`('matematica.py', 'w'):` Abre o arquivo `matematica.py` no modo de escrita `('w'). `
Se o arquivo no existir, ele ser criado. Se o arquivo j existir, seu contedo ser substitudo.
as `f:` A varivel f  um objeto de arquivo que voc usar para manipular o contedo do arquivo.
`f.write("""..."""):` Escreve o contedo dentro das aspas triplas `("""...""")` no arquivo. As aspas triplas permitem que voc escreva strings multilinha em Python, facilitando a escrita de blocos de texto que contm vrias linhas.
